### 2-1. 변수의 이름

case 1. 알파벳 대소문자와 한글도 가능하다.<br> (한글은 가능하지만 한영변환을 자주해야해서 불편하다.)

case 3. 알파벳 + 언더바(_) (가능)

```
a_ = 1
```

case 4. 언더바(_) + 알파벳 (가능))

```
_a = 1
```

```
a b = 6
```
